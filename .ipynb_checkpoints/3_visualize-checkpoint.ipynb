# 4. Visualize Developers vs. Non-Developers

Visualize the mean and spread of the variables used in logistic regression:
- 925-hPa vorticity
- 975-hPa convergence
- 700-hPa specific humidity

# Import statements

```python
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
```

# Load data

```python
years_dev = np.arange(1980, 2022 + 1)
years_nondev_12 = [2000,2002,2003,2004,2005,2006,2007,2009,2010,2012,2014,2015,2016,2017,2018,2019,2020,2021,]
years_nondev_16 = [2001,2002,2003,2004,2005,2007,2009,2011,2012,2014,2015,2016,2017,2018,2019,2020]

dev_list = [
    f"./ERA5_extracted_years/developers_{year}_with_ERA5.nc" for year in years_dev
]
nondev_list = [
    f"./ERA5_extracted_years/non-developer_16pointss_{year}_with_ERA5.nc"
    for year in years_nondev_16
]


dev = xr.open_mfdataset(
    dev_list, combine="nested", concat_dim="track_point", join="outer"
)

nondev = xr.open_mfdataset(
    nondev_list, combine="nested", concat_dim="track_point", join="outer"
)
```

# Plotting function

```python
def plot_timeseries(
    data, var_name, title=None, ylim=None, save=False, savename=None, plot_mean=False
):
    var_dict = {
        "vo_925": "925-hPa vorticity (1/s)",
        "conv_975": "975-hPa convergence (1/s)",
        "q_700": "700-hPa specific humidity (kg/kg)",
    }

    groups = data.groupby("id")
    ids = [id for id, idx in groups.groups.items() if len(idx) >= min_points]

    if len(ids) == 0:
        print(f"No storms have at least {min_points} points.")
        return

    print(f"Using {len(ids)} storms with ≥ {min_points} points")

    rel_hours = np.arange(0, 6 * min_points, 6)

    cmap = plt.get_cmap("jet", len(ids))
    
    plt.figure(figsize=(6, 4))

    all_vals = []

    for i, id in enumerate(ids):
        idx = groups.groups[id][:min_points]
        track = data.isel(track_point=idx)

        vals = np.asarray(track[var_name].values)

        if not plot_mean:
            plt.plot(
                rel_hours, vals, color=cmap(i), linewidth=1.5, alpha=0.7
            )

        all_vals.append(vals)

    all_vals = np.array(all_vals)

    if plot_mean:
        mean_vals = np.mean(all_vals, axis=0)
        p20_vals = np.percentile(all_vals, 20, axis=0)
        p80_vals = np.percentile(all_vals, 80, axis=0)

        plt.plot(rel_hours, mean_vals, color="k", linewidth=2.5)
        plt.fill_between(rel_hours, p20_vals, p80_vals, color="gray", alpha=0.3)

    plt.xlabel("Time since first point (hours)")
    plt.ylabel(var_dict[var_name])
    plt.ylim(ylim)
    plt.xlim(0, 6 * (min_points - 1))
    plt.grid(True, alpha=0.4)
    plt.title(title)
    plt.tight_layout()

    if save:
        plt.savefig("./figures/" + savename)

    plt.show()
```

## Plot

```python
min_points = 16
plot_timeseries(
    dev,
    "vo_925",
    title="Developer: Timeseries Since Seed First Point",
    ylim=[0, 0.002],
    save=False,
    savename=None,
    plot_mean=False,
)
plot_timeseries(
    nondev,
    "vo_925",
    title="Non-Developer: Timeseries Since Seed First Point",
    ylim=[0, 0.002],
    save=False,
    savename=None,
    plot_mean=False,
)
plot_timeseries(
    dev,
    "conv_975",
    title="Developer: Timeseries Since Seed First Point",
    ylim=[-0.0002, 0.0005],
    save=False,
    savename=None,
    plot_mean=False,
)
plot_timeseries(
    nondev,
    "conv_975",
    title="Non-Developer: Timeseries Since Seed First Point",
    ylim=[-0.0002, 0.0005],
    save=False,
    savename=None,
    plot_mean=False,
)
plot_timeseries(
    dev,
    "q_700",
    title="Developer: Timeseries Since Seed First Point",
    ylim=[0, 0.015],
    save=False,
    savename=None,
    plot_mean=False,
)
plot_timeseries(
    nondev,
    "q_700",
    title="Non-Developer: Timeseries Since Seed First Point",
    ylim=[0, 0.015],
    save=False,
    savename=None,
    plot_mean=False,
)
```
